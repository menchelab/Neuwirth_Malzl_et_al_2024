```python
import pandas as pd

data = pd.read_csv(
    '../atacseq/atacseq_counts.tsv',
    sep = '\t'
)

with open('../atacseq/atacseq_counts.tsv', 'r') as countfile:
    headerline = countfile.readline().rstrip()
    raw_header = headerline[1:].split('\t')
    processed_header = [name[1:-1].split('.')[0] for name in raw_header]

    
data.columns = processed_header
data
```

```python
is_canonical = ~data.chr.str.contains('_')
# this is mainly because some of the peaks look like they are in repetitive regions so we filter out high read count peaks
has_resonable_readcount = (data.loc[:, data.columns.str.contains('treg')] < 1500).any(axis = 1)
filtered_data = data.loc[is_canonical & has_resonable_readcount, :]
filtered_data.index = filtered_data[['chr', 'start', 'end']].apply(lambda x: ':'.join([str(i) for i in x.values]), axis = 1)
filtered_data = filtered_data.drop(columns = ['chr', 'start', 'end'])
filtered_data.to_csv('../atacseq/atacseq_counts.filtered.csv')
```
